### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
#Ask BCS showed me how to do the scores.Hygience syntax for drilling into a second level of the library
query = {'scores.Hygiene': {'$eq': 20}}
results = establishments.find(query)
# Use count_documents to display the number of documents in the result
print("Number of documents in result:", establishments.count_documents(query))
# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print("Number of rows in result: ", len(hygiene_df))
# Display the first 10 rows of the DataFrame
hygiene_df.head(10)

#The below table shows 10 establishments with a hygiene score of 20. These are not in any particular order.
```

```python
#The below code functions to determine the restaurants within local range of Penang Flavours, with RatingValue of 5, sorted by lowest hygiene score.
#This code was provided by AskBCS after I had trouble with the above.
degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000
query = {'geocode.latitude': {'$gte':latitude-degree_search, '$lte':latitude+degree_search},
         'geocode.longitude': {'$gte': longitude-degree_search, '$lte': longitude+degree_search},
         'RatingValue': 5}
sort = [('scores.Hygiene', 1)]
limit = 5
# Print the results
pprint(list(establishments.find(query).sort(sort).limit(limit)))
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
# match_query = {'scores.Hygiene': {'$eq': 0}}
# four_group_query = {'$group': {
#                         '_id': "$LocalAuthorityName",
#                         'count':{ '$sum':1}
#                          }
#               }
# sort_query = {'$sort':{'count':-1}}
# pipeline = [match_query,four_group_query, sort_query]
# I was having trouble with this assignment using the above syntax--the AskBCS gave me the below block of code, which worked
match_query=[ 
    {'$match':{'scores.Hygiene':0}},
    {'$group':{'_id':"$LocalAuthorityName",'count':{'$sum':1}}},
    {'$sort':{'count':-1}}
]

results = list(establishments.aggregate(match_query))

# Print the number of documents in the result
print("Number of rows in result: ", len(results))
# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
four_rating_df=pd.DataFrame(results)
# Display the number of rows in the DataFrame
print("Number of rows in result: ", len(four_rating_df))
# Display the first 10 rows of the DataFrame
four_rating_df.head(10)
#There are 55 establishments in each Local Authority area with a hygiene score of 10.
```
